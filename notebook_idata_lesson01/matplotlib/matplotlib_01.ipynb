```python
import pandas

unrate = pandas.read_csv("E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\matplotlib\\UNRATE.csv")
unrate["DATE"] = pandas.to_datetime(unrate["DATE"])
unrate.head(12)
```

```python
import matplotlib.pyplot as plt

# 绘图
plt.plot()
# 显示图形
plt.show()
```

```python
first_twelve = unrate[0:12]
# xy轴数据
plt.plot(first_twelve["DATE"], first_twelve["VALUE"])
plt.show()
```

```python
plt.plot(first_twelve["DATE"], first_twelve["VALUE"])
# 把x轴的tick，旋转45度，避免有的时候tick显示不全
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(first_twelve["DATE"], first_twelve["VALUE"])
plt.xticks(rotation=90)
# xy轴的标签
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
# 折线图的标题
plt.title("Monthly Unemployment Trends,1948")
plt.show()
```
