```python
import pandas
import matplotlib.pyplot as plt
import numpy

unrate = pandas.read_csv("E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\matplotlib\\UNRATE.csv")
unrate["DATE"] = pandas.to_datetime(unrate["DATE"])
```

```python
fig = plt.figure()

# 添加子图，三个参数表示，分成几行几列，然后在第几个位置加一个子图(从1开始)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax4 = fig.add_subplot(2, 2, 4)
plt.show()
```

```python
fig = plt.figure()
# figsize可以设置fig的大小
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# 在子图上绘图
ax1.plot(numpy.random.randint(1, 5, 5), numpy.arange(5))
ax2.plot(numpy.arange(10) * 3, numpy.arange(10))
plt.show()
```

```python
# 在同一个图中，画出两条折线
unrate["MONTH"] = unrate["DATE"].dt.month
fig = plt.figure(figsize=(6, 3))

# 绘制两条折线
plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["VALUE"], c="red")
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["VALUE"], c="blue")

plt.show()
```

```python
# 绘制多条折线
flg = plt.figure(figsize=(10, 6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"], subset["VALUE"], c=colors[i])

plt.show()
```

```python
# 设置线段的颜色的表示
fig = plt.figure(figsize=(10, 6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index: end_index]
    label = str(1948 + i)
    # 设置线段的表示，label属性
    plt.plot(subset["MONTH"], subset["VALUE"], c=colors[i], label=label)
# 设置线段的表示，best是表示默认最佳位置，还有其他的参数可以填，可以用help看还有什么参数
plt.legend(loc="best")

plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index: end_index]
    label = str(1948 + i)
    plt.plot(subset["MONTH"], subset["VALUE"], c=colors[i], label=label)
plt.legend(loc="upper left")
plt.xlabel("Month,Integer")
plt.ylabel("Unemployment Rate,Percent")
plt.title("Monthly Unemployment Trends,1948-1952")

plt.show()
```
