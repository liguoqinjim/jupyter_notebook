```python
# 读取数据
import pandas as pd

titanic = pd.read_csv(
    "E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\Titanic\\titanic_train.csv"
)
# 查看数据大致情况
print(titanic.describe())
titanic[:3]
```

```python
# 处理Age
# 补上缺失值，使用中位数来补上缺失值
mid = titanic["Age"].median()
print(mid)
# 错误的方法
# titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
titanic["Age"] = titanic["Age"].fillna(mid)
```

```python
# 处理Sex
print(titanic["Sex"].unique())

# 使用数值代替字符串，male为0，female为1
titanic.loc[titanic["Sex"] == "male", "Sex"] = 0
titanic.loc[titanic["Sex"] == "female", "Sex"] = 1

titanic[:3]
```

```python
# 处理Embarked
print(titanic["Embarked"].unique())

# 补充缺失值，但是Embarked不是数值型的数据，只能补上3个值里面最多的一个值
titanic["Embarked"] = titanic["Embarked"].fillna("S")
# 使用数值代替字符串
titanic.loc[titanic["Embarked"] == "S", "Embarked"] = 0
titanic.loc[titanic["Embarked"] == "C", "Embarked"] = 1
titanic.loc[titanic["Embarked"] == "Q", "Embarked"] = 2

titanic[:3]
```

```python
# 线性回归
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import KFold

# 要参数建模的特征
predictors = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]

# 创建线性回归
alg = LinearRegression()
# 交叉分析
kf = KFold(titanic.shape[0], n_folds=3, random_state=0)

predictions = []
for train_index, test_index in kf:
    # 交叉分析里面得到的本次的训练集index
    train_data = titanic[predictors].iloc[train_index, :]
    # 交叉分析里面的测试集的index
    train_target = titanic["Survived"].iloc[train_index]

    # 分析
    alg.fit(train_data, train_target)
    test_predictions = alg.predict(titanic[predictors].iloc[test_index, :])
    predictions.append(test_predictions)
```

```python
# 统计线性回归正确率
import numpy as np

predictions = np.concatenate(predictions, axis=0)

predictions[predictions > .5] = 1
predictions[predictions <= .5] = 0

correct = len(predictions[predictions == titanic["Survived"]])
accuracy = correct / len(predictions)
print(accuracy)
```

```python
# 逻辑回归
from sklearn import cross_validation
from sklearn.linear_model import LogisticRegression

alg = LogisticRegression(random_state=0)
scores = cross_validation.cross_val_score(alg, titanic[predictors], titanic["Survived"], cv=3)
print(scores.mean())
```

```python
# 再次调整数据（在现有数据里面）

# SibSp和Parch合并
titanic["FamilySize"] = titanic["SibSp"] + titanic["Parch"]

# Name长度
titanic["NameLength"] = titanic["Name"].apply(lambda x: len(x))
```

```python
# 处理Name，找出里面的title
import re

def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # If the title exists, extract and return it.
    if title_search:
        return title_search.group(1)
    return ""


titles = titanic["Name"].apply(get_title)
print(pd.value_counts(titles))

title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Dr": 5, "Rev": 6, "Major": 7, "Col": 7, "Mlle": 8,
                 "Mme": 8, "Don": 9, "Lady": 10, "Countess": 10, "Jonkheer": 10, "Sir": 9, "Capt": 7, "Ms": 2}
for k, v in title_mapping.items():
    titles[titles == k] = v

# Verify that we converted everything.
print(pd.value_counts(titles))

# Add in the title column.
titanic["Title"] = titles
```

```python
# 画出特征重要性
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

predictors = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked", "FamilySize", "Title", "NameLength"]

selector = SelectKBest(f_classif, k=5)
selector.fit(titanic[predictors], titanic["Survived"])

scores = -np.log10(selector.pvalues_)

sns.set()
plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation="vertical")
plt.show()
```

```python
# 集成算法
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np

# 集成算法里面的算法
algorithms = [
    [GradientBoostingClassifier(random_state=1, n_estimators=25, max_depth=3),
     ["Pclass", "Sex", "Age", "Fare", "Embarked", "FamilySize", "Title", ]],
    [LogisticRegression(random_state=1), ["Pclass", "Sex", "Fare", "FamilySize", "Title", "Age", "Embarked"]]
]

# 交叉验证
kf = KFold(titanic.shape[0], n_folds=3, random_state=1)

predictions = []
for train, test in kf:
    train_target = titanic["Survived"].iloc[train]
    full_test_predictions = []

    for alg, predictors in algorithms:
        alg.fit(titanic[predictors].iloc[train, :], train_target)
        test_predictions = alg.predict_proba(titanic[predictors].iloc[test, :].astype(float))[:, 1]
        full_test_predictions.append(test_predictions)
        
    test_predictions = (full_test_predictions[0] + full_test_predictions[1]) / 2
    test_predictions[test_predictions <= .5] = 0
    test_predictions[test_predictions > .5] = 1
    predictions.append(test_predictions)

# Put all the predictions together into one array.
predictions = np.concatenate(predictions, axis=0)

# Compute accuracy by comparing to the training data.
accuracy = len(predictions[predictions == titanic["Survived"]]) / len(predictions)
print(accuracy)
```
