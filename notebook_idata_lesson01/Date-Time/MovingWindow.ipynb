```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
df = pd.Series(np.random.randn(600), index=pd.date_range("2017-09-15", freq="D", periods=600))
df[:5]
```

```python
# 滑动窗口默认从左边开始滑动，滑动窗口的长度，到数据本身停止
r = df.rolling(window=10)
r
```

```python
print(r.mean())
```

```python
sns.set()
plt.figure(figsize=(18, 8))

df.plot(style="r--")
df.rolling(window=10).mean().plot(style="b")
```
