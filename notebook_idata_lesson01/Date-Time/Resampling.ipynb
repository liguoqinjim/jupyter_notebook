```python
import pandas as pd
import numpy as np

rng = pd.date_range("2017-9-15", periods=90, freq="D")
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts[:5]
```

```python
# 降采样(采样点减少，从90天减少到4个月份)
ts.resample("M").sum()
```

```python
# 降采样
ts.resample("6D").sum()
```

```python
day6Ts = ts.resample("6D").mean()
day6Ts
```

```python
# 升采样(采样点变多)
day6Ts.resample("D").asfreq()

# 会看到很多Nan值，这是升采样会遇到的情况。需要插值填充
# 有三种插值填充
# ffill 空值取前面的值
# bfill 空值取后面的值
# interpolate 线性取值
```

```python
# ffill的参数，表示插值填空几个
day6Ts.resample('D').ffill(2)

# 插值填空2个，可以看到有三个Nan值
```

```python
# bfill
day6Ts.resample('D').bfill(2)

# bfill的插值填空，是从后面开始插值的
```

```python
# 线性插值填空
day6Ts.resample('D').interpolate('linear')
```
