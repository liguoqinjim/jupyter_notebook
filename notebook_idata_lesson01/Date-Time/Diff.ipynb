```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

# 直接读取数据(这里不用教程里面的从网上自动下载数据的方式)
sentiment = pd.read_csv(
    "E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\Date-Time\data\sentiment.csv",
    index_col=0, parse_dates=[0]
)
print(len(sentiment))
sentiment[:5]
```

```python
# 时间过滤
sentiment_short = sentiment.loc["2005":"2016"]
print(len(sentiment_short))
```

```python
# 使用sentiment_short绘图
sns.set()
plt.figure(figsize=(12, 8))
sentiment_short.plot()
plt.title("Consumer Sentiment")
plt.show()
```

```python
# 使用差分法，使数据平稳
# 一阶差分
sentiment_short["diff_1"] = sentiment_short["UMCSENT"].diff(1)
# 二阶差分
sentiment_short["diff_2"] = sentiment_short["diff_1"].diff(1)

# 绘图
sns.set()
plt.figure(figsize=(20, 13))
sentiment_short.plot(subplots=True)
```
