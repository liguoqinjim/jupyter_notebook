```python
% matplotlib inline
from skimage import io
from sklearn.cluster import KMeans
import numpy as np

# 显示原图
image = io.imread("E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\KMEANS\data\\test2.jpg")
io.imshow(image)
io.show()

print("image.shape=", image.shape)
rows = image.shape[0]
cols = image.shape[1]

# reshape
image = image.reshape(rows * cols, 3)
# n_clusters就是K参数，就是簇
kmeans = KMeans(n_clusters=128, n_init=10, max_iter=200)
kmeans.fit(image)

clusters = np.asarray(kmeans.cluster_centers_, dtype=np.uint8)
labels = np.asarray(kmeans.labels_, dtype=np.uint8)
labels = labels.reshape(rows, cols)

print("clusters.shape=", clusters.shape)

np.save("E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\KMEANS\data\codebook_test.npy", clusters)
io.imsave("E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\KMEANS\data\compressed_test.jpg", labels)
```
