```python
# 读取数据
import pandas as pd

beer = pd.read_csv(
    "E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\KMEANS-DBSCAN\data.txt",
    sep=" "
)
beer[:5]
```

```python
x = beer[["calories", "sodium", "alcohol", "cost"]]
x[:5]
```

```python
# KMEANS
from sklearn.cluster import KMeans

# 分成3簇
km = KMeans(n_clusters=3).fit(x)
# 分成2簇
km2 = KMeans(n_clusters=2).fit(x)

# 聚类之后的结果
print(km.labels_)

# 赋值到beer
beer["cluster"] = km.labels_
beer["cluster2"] = km2.labels_
beer.sort_values("cluster")
```

```python
% matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style(style="whitegrid")

plt.rcParams['font.size'] = 14
colors = np.array(['red', 'green', 'blue', 'yellow'])
plt.scatter(beer["calories"], beer["alcohol"], c=colors[beer["cluster"]])

# km的聚类结果绘制
centers = beer.groupby("cluster").mean().reset_index()
plt.scatter(centers.calories, centers.alcohol, linewidths=3, marker='+', s=300, c='black')

plt.xlabel("Calories")
plt.ylabel("Alcohol")
```

```python
from pandas.plotting import scatter_matrix

# 三个簇的结果
scatter_matrix(beer[["calories", "sodium", "alcohol", "cost"]], s=100, alpha=1, c=colors[beer["cluster"]],
               figsize=(15, 15))
plt.suptitle("With 3 centroids initialized")
```

```python
# 两个簇的结果
scatter_matrix(beer[["calories", "sodium", "alcohol", "cost"]], s=100, alpha=1, c=colors[beer["cluster2"]],
               figsize=(15, 15))
plt.suptitle("With 2 centroids initialized")
```

```python
# 标准化数据
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
print(x_scaled[:5])

km = KMeans(n_clusters=3).fit(x_scaled)
beer["scaled_cluster"] = km.labels_
beer.sort_values("scaled_cluster")
```

```python
# 数据标准化之后三个簇的结果
scatter_matrix(beer[["calories", "sodium", "alcohol", "cost"]], s=100, alpha=1, c=colors[beer["scaled_cluster"]],
               figsize=(15, 15))
plt.suptitle("With 3 centroids initialized")
```

```python
# 计算轮廓系统，评估聚类的结果
from sklearn import metrics

score_scaled = metrics.silhouette_score(x, beer["scaled_cluster"])
score = metrics.silhouette_score(x, beer["cluster"])
print("score_scaled=", score_scaled)
print("score=", score)
```

```python
# 测试参数K设置为多少比较好，也就是设置为多少簇
scores = []
for k in range(2, 20):
    labels = KMeans(n_clusters=k).fit(x).labels_
    score = metrics.silhouette_score(x, labels)
    scores.append(score)

scores
```

```python
# 绘制scores的值
plt.plot(list(range(2, 20)), scores)
plt.xlabel("Number of Clusters Initialized")
plt.ylabel("Sihouette Score")
```

```python
# DBSCAN聚类
from sklearn.cluster import DBSCAN

# eps是半径，min_samples半径内点的个数
dbscan = DBSCAN(eps=10, min_samples=2).fit(x)
labels = dbscan.labels_
beer["cluster_dbscan"] = labels
beer.sort_values('cluster_dbscan')
```

```python
# 绘图
pd.scatter_matrix(x, c=colors[beer["cluster_dbscan"]], figsize=(15, 15), s=100)
plt.suptitle("DBSCAN")
```

```python
# 使用轮廓函数查看DBSCAN的效果
score = metrics.silhouette_score(x, beer["cluster_dbscan"])
print("score=", score)
```

```python
# 查看哪个参数比较好

scores = []
for eps in range(8, 12, 1):
    for min_samples in range(2, 5):
        # print(eps, min_samples)
        label = DBSCAN(eps=10, min_samples=min_samples).fit(x).labels_
        score = metrics.silhouette_score(x, label)
        scores.append(score)

scores

# 注意：要是label的值都是-1的话，也就是都是离群点的话，会报错
```
