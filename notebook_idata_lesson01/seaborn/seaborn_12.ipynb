```python
% matplotlib inline
import matplotlib.pyplot as plt
import numpy
import seaborn as sns

numpy.random.seed(0)
sns.set_style("whitegrid")
```

```python
uniform_data = numpy.random.rand(3, 3)
print(uniform_data)
# 绘制热度图
sns.heatmap(uniform_data)
```

```python
# 设置最小值和最大值，vmin和vmax
sns.heatmap(uniform_data, vmin=0.2, vmax=0.5)
```

```python
normal_data = numpy.random.randn(3, 3)
print(normal_data)
# 中间值是0
sns.heatmap(normal_data, center=0)
```

```python
# 加载数据
flights = sns.load_dataset("flights")
flights[:3]
```

```python
# year为x轴，month为y轴，passengers为对应的值
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights)
```

```python
# 在热度图中显示出值，fmt为值的数据格式，annot为True的时候显示
sns.heatmap(flights, annot=True, fmt="d")
```

```python
# linewidths设置分割线的宽度
sns.heatmap(flights, linewidths=.5)
```

```python
# 使用其他的颜色
sns.heatmap(flights, cmap="YlGnBu")
```

```python
# 是否显示颜色条，cbar为False的时候不显示
sns.heatmap(flights, cbar=False)
```
