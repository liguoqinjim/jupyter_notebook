```python
% matplotlib inline
import numpy
import pandas
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns

sns.set(color_codes=True)
numpy.random.seed(sum(map(ord, "distributions")))
```

```python
# 直方图(柱形图)
x = numpy.random.normal(size=100)
sns.distplot(x, kde=False)
```

```python
# 直方图，自定义bins
sns.distplot(x, bins=20, kde=False)
```

```python
# 直方图，查看数据分布情况
x = numpy.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma)
```

```python
# 根据均值和协方差生成数据
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = numpy.random.multivariate_normal(mean, cov, 200)
df = pandas.DataFrame(data, columns=["x", "y"])
df[:5]
```

```python
# 散点图，方便观测两个变量之间的关系
sns.jointplot(x="x", y="y", data=df)
```

```python
# 生成1000个数据，当数据多的时候，不方便用散点图，而是用这种图，kind是hex
# 可以表示一个区域的数据的是多还是少
x, y = numpy.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k")
```
