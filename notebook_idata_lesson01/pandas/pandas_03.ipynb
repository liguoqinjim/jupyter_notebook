```python
import pandas
food_info = pandas.read_csv("E:\Workspace\jupyter_notebook\\notebook_idata_lesson01\pandas\\food_info.csv")
```

```python
food_info["Sodium_(mg)"].head(3)
# 排序,按照Sodium_(mg)从小到大排序,inplace为True的时候，表示就排序本身，而不是返回一个新的排序过的结果
food_info.sort_values("Sodium_(mg)", inplace=True)
print(food_info["Sodium_(mg)"])
# print(food_info)
```

```python
# 从大到小排序，ascending为False的时候表示从大到小排序
food_info.sort_values("Sodium_(mg)", inplace=True, ascending=False)
print(food_info["Sodium_(mg)"])
```
